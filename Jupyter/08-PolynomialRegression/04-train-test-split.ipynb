```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
np.random.seed(666)
x = np.random.uniform(-3, 3, size=100)
X = x.reshape(-1, 1)
y = 0.5*x**2 + x + 2 + np.random.normal(0, 1, size=100)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler

def PolynomialRegression(degree):
	return Pipeline([
		("poly", PolynomialFeatures(degree = degree)),
		("std_scaler", StandardScaler()), 
		("lin_reg", LinearRegression())
		])
```

```python
poly_reg2 = PolynomialRegression(2)
```

```python
poly_reg2.fit(X_train, y_train)
```

```python
y_predict2 = poly_reg2.predict(X_test)
```

```python
#plt.scatter(x, y)
#plt.plot(np.array(np.sort(x)), y_predict2[np.argsort(x)], color='r')
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_predict2)
```

```python
poly_reg10 = PolynomialRegression(10)
poly_reg10.fit(X_train, y_train)
y_predict10 = poly_reg10.predict(X_test)
plt.scatter(x, y)
#plt.plot(np.array(np.sort(x)), y_predict2[np.argsort(x)], color='b')
#plt.plot(np.array(np.sort(x)), y_predict10[np.argsort(x)], color='r')
mean_squared_error(y_test, y_predict10)
```

```python
poly_reg100 = PolynomialRegression(100)
poly_reg100.fit(X_train, y_train)
y_predict100 = poly_reg100.predict(X_test)
plt.scatter(x, y)
#plt.plot(np.array(np.sort(x)), y_predict2[np.argsort(x)], color='b')
#plt.plot(np.array(np.sort(x)), y_predict100[np.argsort(x)], color='r')

mean_squared_error(y_test, y_predict100)
```
