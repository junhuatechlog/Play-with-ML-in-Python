```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-np.pi, np.pi, 101)
y = np.sin(x)+np.sin(3*x)/3.0
fig = plt.figure()
ax = fig.add_subplot(121)
ax.plot(x, y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('A simple plot')
```

```python
x = np.linspace(0, 5, 101)
y1 = 1.0/(x+1.0)
y2 = np.exp(-x)
y3 = np.exp(-0.1*x**2)
y4 = np.exp(-5*x**2)

fig = plt.figure() # 
ax1 = fig.add_subplot(221) #
ax1.plot(x, y1)
ax1.set_xlabel('x')
ax1.set_ylabel('y1')
ax1.set_title('A simple plot')#
ax2 = fig.add_subplot(222)# (2, 2, 2)
ax2.plot(x, y2)
ax2.set_xlabel('x')
ax2.set_ylabel('y2')
ax2.set_title('A simple plot')
ax3 = fig.add_subplot(223)
ax3.plot(x, y3)
ax3.set_xlabel('x')
ax3.set_ylabel('y3')
ax3.set_title('A simple plot')
ax4 = fig.add_subplot(224)
ax4.plot(x, y4)
ax4.set_xlabel('x')
ax4.set_ylabel('y4')
ax4.set_title('A simple plot')
fig.tight_layout()
fig.suptitle('Various decay functions') #
```
