```python
def gaussian(x, l):
    gamma = 1.0
    return np.exp(-gamma * (x - l)**2)
```

```python
l1, l2 = -1, 1

X_new = np.empty((len(x), 2))
for i, data in enumerate(x):
    X_new[i, 0] = gaussian(data, l1)
    X_new[i, 1] = gaussian(data, l2)
```

```python
X_new
```

```python
plt.scatter(X_new[y==0, 0], X_new[y==0, 1])
plt.scatter(X_new[y==1, 0], X_new[y==1, 1])
```
