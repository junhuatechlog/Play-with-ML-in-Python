```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
data = datasets.load_iris()
X = data.data[:, 2:]
y = data.target
from sklearn.tree import DecisionTreeClassifier
dtc_clf = DecisionTreeClassifier(max_depth = 2, criterion='entropy')
dtc_clf.fit(X, y)

def plot_decision_boundary(model, axis):
    x0, x1 = np.meshgrid(
        np.linspace(axis[0], axis[1], int((axis[1] - axis[0])*100)).reshape(-1, 1), 
        np.linspace(axis[2], axis[3], int((axis[3] - axis[2])*100)).reshape(-1, 1)
    )
    X_new = np.c_[x0.ravel(), x1.ravel()]

    y_predict = model.predict(X_new)
    zz = y_predict.reshape(x0.shape)

    from matplotlib.colors import ListedColormap
    custom_cmap = ListedColormap(['#EF9A9A', '#EFF59D', '#90CAF9'])

    plt.contourf(x0, x1, zz, cmap=custom_cmap)
    
plot_decision_boundary(dtc_clf, [0.5, 7.5, 0, 3])
plt.scatter(X[y==0, 0], X[y==0, 1])
plt.scatter(X[y==1, 0], X[y==1, 1])
plt.scatter(X[y==2, 0], X[y==2, 1])
plt.title("Decision Tree")
```

```python
X_new = np.delete(X, 138, axis = 0)# 把第138个样本删掉
y_new = np.delete(y, 138)
```

```python
tree_clf= DecisionTreeClassifier(max_depth = 2, criterion='entropy', random_state=111)
```

```python
tree_clf.fit(X_new, y_new)
```

```python
plot_decision_boundary(tree_clf, [0.5, 7.5, 0, 3])
plt.scatter(X_new[y_new==0, 0], X_new[y_new==0, 1])
plt.scatter(X_new[y_new==1, 0], X_new[y_new==1, 1])
plt.scatter(X_new[y_new==2, 0], X_new[y_new==2, 1])
plt.title("Decision Tree")
```

```python
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
data= pd.read_csv("boston_house_prices.csv", skiprows=[0])
array = data.values
```

```python
array.shape
```

```python
X = array[:, :13]
y = array[:, 13]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 666)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr = DecisionTreeRegressor()
```

```python
dtr.fit(X_train, y_train)
```

```python
dtr.score(X_test, y_test)
```

```python
dtr.score(X_train, y_train)
```

```python
from sklearn.tree import export_graphviz
```

```python
export_graphviz(dtr, out_file="iris_tree.dot", rounded=True, filled=True)
```
