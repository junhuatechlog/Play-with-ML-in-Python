### 简单线性回归 - x 是一个向量，不是矩阵

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 3, 2, 3, 5])
```

```python
plt.scatter(x, y, color = 'g')
```

```python
x_mean = np.mean(x)
```

```python
x_mean
```

```python
y_mean = np.mean(y)
```

```python
y_mean
```

```python
np.sum((x - x_mean)*(y - y_mean))
```

```python
np.sum((x-x_mean)**2)
```

```python
a = 8/10
```

```python
a
```

```python
b = y_mean - a*x_mean
```

```python
b
```

```python
y_hat = x*a + b
```

```python
y_hat
```

```python
plt.scatter(x, y, color='g')
plt.plot(x, y_hat, color='r')
```

```python
(x - x_mean).dot(y - y_mean)
```

```python
num = 0.0
d = 0.0
for x_i, y_i in zip(x, y):
    num += (x_i - x_mean) *(y_i - y_mean)
    d += (x_i - x_mean)**2
```

```python
num
```

```python
d
```

```python
d
```
