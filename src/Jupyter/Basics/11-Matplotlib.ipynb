```python
X = np.linspace(0, 10, 100)
plt.xlim(-5, 15)
plt.ylim(0, 1.5)
plt.axis([-1, 11, -2, 2])
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(X, y, color = "red", linestyle = "--", label="sin(x)")
Z = np.cos(X)
plt.plot(X, Z, color  = "brown", linestyle = "-.", label="cos(x)")
plt.title("test")
```

```python
X = np.linspace(0, 10, 100)
plt.xlim(-5, 15)
plt.ylim(0, 1.5)
plt.axis([-1, 11, -2, 2])
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.scatter(X, y, color = "red", linestyle = "--", label="sin(x)")
Z = np.cos(X)
plt.plot(X, Z, color  = "brown", linestyle = "-.", label="cos(x)")
plt.title("test")
```
